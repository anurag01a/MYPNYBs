 A companyâ€™s sales data of different items in various regions is given in the following table (in 
%age). Whereas the target to do business in each region is given in the last row as TARGET. 
Using this data plot different type of graphs using matplotlib package in Python, giving title, 
target with percentage and actual percentage data on x-axis and y-axis respectively. Also 
indicate each item details using legend within the graph figure.
 

image.png

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = [41.75, 36.50, 31.94, 31.10, 29.35]
y = np.array([[34.35, 24.87, 20.50, 17.64, 15.50],[41.59, 35.87, 32.99, 30.93, 29.28],[41.64, 35.77, 32.74, 30.48, 28.42],[42.18, 36.20, 33.16, 30.92, 29.22],[42.23, 36.19, 33.15, 30.87, 28.83],[40.43, 34.41, 31.27, 28.80, 26.55],[40.28, 34.29, 31.16, 28.73, 26.43]])
```

```
y
```

#**Line Chart**

```
C=0
plt.figure(figsize=(10,10))
for a in y:
  C=C+1
  plt.grid(color='g',linestyle='-',linewidth=1)
  ax = plt.subplot(111)
  ax.plot(x, a, label='Item{}'.format(C)) 
  plt.title('Sales Data with Legend inside')
  plt.xlabel('Target with Percentage')
  plt.ylabel('Actual Percentage Data')
  ax.legend()  
  plt.plot(x,a,marker ='o')
```

#**Scatter Plot**

```
color = ["blue", "yellow", "pink", "green", "brown", "red", "violet"]
mark = ["+","*",".","p","s","^","<"]
i=0
plt.figure(figsize=(10,10))
for a in y:  
  plt.scatter(x, a, c =color[i], 
              linewidths = 2, 
 
              marker = mark[i],  
 
              edgecolor =color[-i],  
 
              s = 200)  
  plt.grid(color='g')  
  ax = plt.subplot(111)  
  ax.scatter(x, a, label='Item{}'.format(i+1),marker = mark[i]) 
  plt.title('Sales Data with Legend inside')
  ax.legend()
  i = i+1
plt.title('Scatter plot ')
plt.xlabel('Target with Percentage')
plt.ylabel('Actual Percentage Data')
plt.show()
```

#**Bar Plot**

```
i = 0
plt.figure(figsize=(10,10))
for a in y:  
  plt.grid(color='g')
  ax = plt.subplot(111)
  ax.bar(x, a, label='Item{}'.format(i+1)) 
  plt.title('Sales Data with Legend inside')
  plt.xlabel('Target with Percentage')
  plt.ylabel('Actual Percentage Data')
  ax.legend() 
  i=i+1
plt.title('Bar plot ')
plt.show()
```

#**Box Plot**

```
plt.figure(figsize=(10,10))
for a in y:    
  plt.boxplot(a) 
plt.title("Sales Targets(boxplot Plot)",fontsize=25)
plt.ylabel("Original values(in %)-------->",fontsize=20)
plt.tick_params(labelsize=15)
plt.show()
```
